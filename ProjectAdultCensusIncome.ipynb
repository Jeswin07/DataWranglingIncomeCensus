<h4 style='color:tomato'>Data Cleaning</h4>

<h5>Before dive deeper into visualizations and insights, we need to clean our dataset</h5>
<h5>This involves handling duplicates, missing values, correcting data types if needed and ensuring our data is ready for analysis</h5>

**Handling duplicate rows**

The count of total duplicate rows are found, and those data are dropped.

```python
# find total count of duplicate row

income_census.duplicated().sum()
```

```python
# drop all 24 duplicate rows

income_census.drop_duplicates(keep='first')
```

**Missing Value Percentage Analysis**

Instead of just counting the number of missing values, we will calculate the percentage of missing values in each column.
This will help us decide whether to drop the column or handle it using imputation 

```python
# to find the missing value percentage in each column

missing_percentage = (income_census.isnull().sum()) / (len(income_census)) * 100

missing_percentage
```

```python
# a dataframe is created to store columns with null values and sorted with highest missing values on top

missing_df = pd.DataFrame({"Columns": income_census.columns, "MissingPercentage": missing_percentage})
missing_df = missing_df[missing_df["MissingPercentage"]>0].sort_values(by="MissingPercentage", ascending=False)

missing_df
```

**Observation:**

- The `occupation` and `workclass` column has about ~5.6% missing entries.
- The `native.country` column has only about ~1.7% missing entries.

Since all 3 columns contain categorical values, the best option will be to fill using mode.

```python
# filling the columns with mode

income_census["occupation"] = income_census["occupation"].fillna(income_census["occupation"].mode()[0])
income_census["workclass"] = income_census["workclass"].fillna(income_census["workclass"].mode()[0])
income_census["native.country"] = income_census["native.country"].fillna(income_census["native.country"].mode()[0])

income_census.isnull().sum()
```

**Capitalizing `sex` column**

To ensure consistency in text values, we capitalized first letter in the 'SEX' column.

```python
# capitalize

income_census["sex"] = income_census["sex"].str.capitalize()
```

**Renaming Columns for clarity**

Renaming `fnlwgt` to `final_weight` for better readability and understanding.

```python
# rename

income_census.rename(columns={"fnlwgt": "final_weight"}, inplace=True)

income_census
```

<h4 style='color:tomato'>Feature Engineering</h4>

<h5>In this section, we'll extract new features from the existing data to gain more insights and improve the quality of analysis.</h5>

**Extracting `net_capital` from `capital.gain` and `capital.loss`**

We can combine `capital.gain` and `capital.loss` to find `net_capital`.
Equation: (`capital.gain` - `capital.loss`)

```python
# create a new column for net capital

income_census["net_capital"] = income_census["capital.gain"] - income_census["capital.loss"]
```

**Bin `age` into categories**

We will categorise `age` into bin of specific age ranges:
- Youth (17-25)
- Young Adult (26-40) 
- Middle Age (41-60) 
- Senior (>60)

```python
# categorise age into bins

income_census["age_bin"] = pd.cut(income_census["age"], bins=[0,20,40,60,100], labels=["17-25", "26-40", "41-60", "60+"], right=True)
```

**Bin `hours.per.week` into categories**

We will categorise `hours.per.week` into bin of specific age ranges:
- Less than 20
- 20-40
- 40-60
- 60+

```python
income_census["hours_bin"] = pd.cut(income_census["hours.per.week"], bins=[0,20,40,60,100], labels=["0-20", "21-40", "41-60", "60+"], right=True)
```

**Simplify `education`**

Simplify education into broader categories

```python
# simplify and assign school and associates
income_census['education_simplified'] = income_census['education'].replace([
    'Preschool', '1st-4th', '5th-6th', '7th-8th', '9th', '10th', '11th', '12th'
], 'School')

income_census['education_simplified'] = income_census['education_simplified'].replace([
    'Assoc-voc', 'Assoc-acdm'
], 'Associates')
```

**Encoding the `income` column**

We will convert the categorical `income` column into numerical values:
- 0 for <=50K
- 1 for >50K

This step is part of feature engineering and helps in numerical computations, such as correlation analysis and machine learning models.

```python
income_census["income_encoded"] = income_census["income"].map({"<=50K": 0, ">50K": 1})
```

**Encoding the `sex` column**

We will convert the categorical `Sex` column into numerical values:
- 0 for male
- 1 for female

This step is part of feature engineering and helps in numerical computations, such as correlation analysis and machine learning models.

```python
# create new column with encoded sex

income_census["sex_encoded"] = income_census["sex"].map({"Male": 0, "Female": 1})
```

<h5>Now that the dataset is cleaned and preprocessed, we'll explore it through visualizations to understand patterns and relationships between features like income, age, sex, family status, working hours and education.</h5>

1. **Gender wise income distribution**

Let's count the income distribution based on gender and visualize the results.

```python
# find count of income category

income_count = income_census["income"].value_counts()

income_count
```

```python

plt.figure(figsize=(6,4))
plt.bar(["Less than 50k", "More than 50k"], income_count)
plt.title("Income balance")
plt.xlabel("Category")
plt.ylabel("Count")
plt.grid(axis = "y", linestyle = "--", alpha = 0.6)
plt.show()
```

2. **Count by age and hours worked**

We can create a subplot with the count of Age and Hours worked per week.

```python

fig, axs = plt.subplots(1,2, figsize=(8,4))
income_census["age_bin"].value_counts(sort=False).plot(kind="bar", ax=axs[0])
axs[0].set_title("Age")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Count")
axs[0].grid(axis="y", linestyle="--", alpha=0.6)

income_census["hours_bin"].value_counts(sort=False).plot(kind="bar", ax=axs[1])
axs[1].set_title("Hours per week")
axs[1].set_xlabel("Hours")
axs[1].set_ylabel("Count")
axs[1].grid(axis="y", linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```

**Observation:**
- Majority of the candidates belong to the Young adult (26-40 yo) category.
- Majority of the candidates work for 21-40 hours a week, some others clocking over 60 hours.

3. **Income by sex**

Let's analysis the gap between Male and Female income.

4. **Work hour distribution**

Analyse the work hour standards with box plot.

5. **Occupational status of high income earners**

Create a pie chart to analyse the occupational position of candidates earning above 50K.

```python
high_income = income_census[income_census["income"] == "<=50K"]
occ_count = high_income["occupation"].value_counts()
occ_count = occ_count[occ_count > 1]
total = occ_count.sum()


plt.figure(figsize=(5,5))
plt.title("Occupation (>=50)")
plt.pie(occ_count, labels=[label if (val / total * 100) >= 1 else '' for label, val in zip(occ_count.index, occ_count)], autopct=lambda pct: f'{pct:.1f}%' if pct >= 1 else '')
plt.show()
```

**Observation:**
The highest earning occupation majorly consists of exec-managers, sale, craft-repair, adm-clerks, prof-speciality etc. It is well distributed.

Using scatter plot, lets analyse the relation between age, education and income.

7. **Multivariate Analysis of education, hours per week and income**

Its done by pivoting the education and hours per week by income category, and then subplotting as Heatmaps. 

```python
# mean of hours worked groupedby education
avg_ed = income_census.groupby("education_simplified")["hours.per.week"].mean()

avg_ed
```

```python
# create pivot table of people with income <=50
income_less50k = income_census[income_census["income"] == "<=50K"].pivot_table(index="education_simplified", columns="hours_bin", aggfunc="size",  observed=False)

income_less50k
```

```python
# create pivot table of people with income >50
income_more50k = income_census[income_census["income"] == ">50K"].pivot_table(index="education_simplified", columns="hours_bin", aggfunc="size", observed=False)

income_more50k
```

```python
# Assigning labels to the pivot tables
all_education_labels = ['School', 'HS-grad', 'Some-college', 'Associates', 'Bachelors', 'Prof-school', 'Masters', 'Doctorate']
all_hours_labels = ["0-20", "21-40", "41-60", "60+"]

income_less50k = income_less50k.reindex(index=all_education_labels, columns=all_hours_labels).fillna(0)
income_more50k = income_more50k.reindex(index=all_education_labels, columns=all_hours_labels).fillna(0)

```

```python
# Heatmap plotting
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(16, 7)) 

sns.heatmap(income_less50k, annot=True, fmt=".0f", cmap='Blues', linewidths=.5, ax=axes[0])
axes[0].set_title('Count of <=50K Income by Education & Hours per Week')
axes[0].set_xlabel('Hours per Week Group')
axes[0].set_ylabel('Education Level')


sns.heatmap(income_more50k, annot=True, fmt=".0f", cmap='Greens', linewidths=.5, ax=axes[1])
axes[1].set_title('Count of >50K Income by Education & Hours per Week')
axes[1].set_xlabel('Hours per Week Group')
axes[1].set_ylabel('Education Level') 


plt.suptitle('Joint Distribution of Education and Work Hours by Income Level', fontsize=16, y=1.02)

plt.show()
```

**Observation**

- People with higher education levels like Bachelor's, Master's, and Doctorate are mostly in the higher income group (earning more than 50K).
- Those with lower education levels such as School, High School graduation, or Some College tend to fall in the lower income group (earning 50K or less).
- Across both income groups, most individuals work between 40 to 60 hours per week.
- Simply working more hours doesnâ€™t guarantee higher income - education level seems to have a bigger impact.
- It's rare to see people with only school or high school education earning above 50K, even if they work full-time.
